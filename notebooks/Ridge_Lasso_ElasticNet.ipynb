```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header = 1)
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df.head()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.loc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.loc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change the required datatype to integer

```python
df.head()
```

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

## Changing the other datatypes to float

```python
objects = [feature for feature in df.columns if df[feature].dtype=='O']
```

```python
objects
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fire_cleaned_dataset.csv',index=False)
```

## Exploratory Data Analysis

```python
# drop day, month, year
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
# category in classes
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
# plot density plot for all the features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Percentage of pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
percentage
```

```python
# Plotting the pie chart
labels = ['fire','not fire']
plt.figure(figsize=(17,7))
plt.pie(percentage,labels=labels,autopct='%1.1f%%')
plt.title('pie chart for classes')
plt.show()
```
